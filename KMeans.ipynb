```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import copy
import os
from sklearn.cluster import KMeans

def visualize(im1, im2, k):
	# displays two images
    im1 = im1.astype('uint8')
    #im2 = im2.astype('uint8')
    f = plt.figure()
    f.add_subplot(1,2, 1)
    plt.imshow(im1,cmap='gray')
    plt.axis('off')
    plt.title('Original')
    f.add_subplot(1,2, 2)
    plt.imshow(im2,cmap='gray')
    plt.axis('off')
    plt.title('Clusters: '+str(k))
    plt.savefig('../Figures/k_means_'+str(k)+'.jpg')
    plt.show()
    return None

def MSE(Im1, Im2):
	# computes error
	Diff_Im = Im2-Im1
	Diff_Im = np.power(Diff_Im, 2)
	Diff_Im = np.sum(Diff_Im)
	Diff_Im = np.sqrt(Diff_Im)
	sum_diff = np.sum(np.sum(Diff_Im))
	avg_error = sum_diff / float(Im1.shape[0]*Im2.shape[1])
	return avg_error
```

```python
james = img.reshape(-1,3)
james.shape
```

```python
ks = [2, 5, 10, 25, 50, 100, 200]
```

```python
means = []
for k in ks:
    kmeans = KMeans(random_state=0,n_clusters=k).fit(james)
    labels = kmeans.labels_
    clusters = kmeans.cluster_centers_
    recon = reconstruct(labels,clusters)
    visualize(img,recon,k)
    means.append(MSE(img,recon))
```
